1. Create a figure object called fig using plt.figure() , Use add_axes to add an axis to the figure
canvas at [0,0,1,1]. Call this new axis ax.

```python
import matplotlib.pyplot as plt
import pandas as pd
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
ax.plot(x, y)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Plot on Custom Axis')
plt.show()
```

2.Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and
[0.2,0.5,.2,.2] respectively.

```python
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
ax1.plot(x1, y1, label='ax1')

x2 = [1, 2, 3, 4, 5]
y2 = [10, 8, 6, 4, 2]
ax2.plot(x2, y2, color='green', label='ax2')
ax1.set_xlabel('X-axis (ax1)')
ax1.set_ylabel('Y-axis (ax1)')
ax1.set_title('Plot on ax1')
ax1.legend()

ax2.set_xlabel('X-axis (ax2)')
ax2.set_ylabel('Y-axis (ax2)')
ax2.set_title('Plot on ax2')
ax2.legend()

plt.show()
```

4.Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

file_path = 'data.txt'  
x_values = []
y_values = []

with open(file_path, 'r') as file:
    for line in file:
        x, y = line.strip().split(',')
        x_values.append(int(x))
        y_values.append(int(y))

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, 'red', label='Data points')  

plt.xlabel('X Axis') 
plt.ylabel('Y Axis')
plt.title('Line Plot from Data in Text File') 
plt.legend()

plt.show()
```

7.Create plot usingsubplots(),scatter, and some other methods on theAxesobject:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30) 
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(14, 5))  

axs[0].scatter(x, y1, color='pink', label='sin(x)')
axs[0].set_title('Scatter Plot')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].legend()

axs[1].plot(x, y2, linestyle='-', marker='o', color='green', label='cos(x)')
axs[1].set_title('Line Plot')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].legend()

plt.tight_layout()

plt.show()
```

8.Generate random data for x and y and create a scatter plot using Matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)  
num_points = 50
x_positions = np.random.rand(num_points) * 10  
y_positions = np.random.rand(num_points) * 20  

plt.scatter(x_positions, y_positions, color='blue', marker='o', label='Random Data')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Scatter Plot')

plt.legend()

plt.show()
```

10.Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of
the data.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='violet', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Randomly Generated Data')

plt.show()
```

11.Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create
your own.

```python
import matplotlib.pyplot as plt
import pandas as pd

categories = ['Rent', 'Transport', 'Food', 'Entertainment', 'Utilities']
expenses = [1000, 600, 300, 400, 200]

plt.pie(expenses, labels=categories, autopct='%1.1f%%', colors=['violet', 'yellow', 'pink', 'green', 'skyblue'], startangle=90)

plt.title('Expense Distribution in Budget')

plt.show()
```
